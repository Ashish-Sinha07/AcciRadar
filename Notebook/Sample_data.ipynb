```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Load data
df = pd.read_csv(r"D:\#Great Learning\Capstone Project\Datasets\Traffic_Crashes_-_Crashes.csv")
```

```python
df['strata'] = (
    df['MOST_SEVERE_INJURY'].astype(str) + '_' + 
    df['WEATHER_CONDITION'].astype(str) + '_' +
    df['LIGHTING_CONDITION'].astype(str)+ '_' +
    df['INJURIES_TOTAL'].astype(str) + '_' + 
    df['CRASH_TYPE'].astype(str) 
 )
```

```python

# Count category frequency
strata_counts = df['strata'].value_counts()

# Keep only strata groups appearing at least 2 times
valid_strata = strata_counts[strata_counts >= 2].index
df_filtered = df[df['strata'].isin(valid_strata)].copy()

print("Original Rows:", len(df))
print("Filtered Rows:", len(df_filtered))

# Now stratified sampling works
df_sample, _ = train_test_split(
    df_filtered,
    train_size=300000,
    stratify=df_filtered['strata'],
    random_state=42
)

df_sample = df_sample.drop('strata', axis=1)
df_sample.to_csv(r'D:\#Great Learning\Capstone Project\Datasets\crash_data.csv', index=False)

print("Sampling Completed Successfully!")
```
